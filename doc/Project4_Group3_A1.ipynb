## **A1: Learning Fair Representations**

# 1.Introduction

We implemented a learning algorithm for fair classification, aiming to achieve both group fairness and individual fairness in classification tasks. Group fairness ensures that the proportion of positive classification within a protected group is similar to the overall population, while individual fairness ensures similar treatment for similar individuals. The method defines fairness as an optimization problem, seeking a good data representation that encodes information effectively while obfuscating any membership information about the protected group.

In LFR, individuals are represented as data points mapping to a probability distribution in a new representation space. The goal is to discard information that identifies whether an individual belongs to the protected subgroup while retaining other information as much as possible. The model utilizes prototypes in the representation space and optimizes these representations to ensure accurate classification tasks. The learning process of LFR involves minimizing an objective function containing terms enforcing statistical parity, retaining information, and achieving accurate classification predictions. The model also learns a restricted form of a distance function as part of the intermediate representation.

```python
import time
import numpy as np
import scipy.optimize as optim
from sklearn.metrics import accuracy_score
```

```python
def d(X, v, alpha):
    N = X.shape[0]
    D = X.shape[1]
    K = len(v)
    dist = np.zeros((N, K))
    for n in range(N):
        for k in range(K):
            for d in range(D):
                dist[n, k] += alpha[d]*(X.iloc[n][d] - v[k, d])**2

    return dist
```

```python
def M_nk(dist, k):
  import numpy as np
  N, K = dist.shape
  numerator = np.exp(-1*dist)
  denominator = np.sum(numerator, axis=1).reshape(-1, 1)
  M_nk = numerator / denominator
  return M_nk
```

```python
def M_k(M_nk):
  import numpy as np
  N, K = M_nk.shape
  M_k = np.sum(M_nk, axis=0) / N
  return M_k
```

```python
def xn_hat(X, M_nk, v):
  import numpy as np
  N,K = M_nk.shape
  D = X.shape[1]
  xn_hat = np.zeros((N, D))
  for n in range(N):
      for d in range(D):
          for k in range(K):
              xn_hat[n, d] += M_nk[n, k]*v[k, d]
  return xn_hat
```

```python
def lx(X, M_nk, xn_hat):
  N,K = M_nk.shape
  D = X.shape[1]
  lx = 0
  for n in range(N):
    for d in range(D):
      lx += (X.iloc[n][d] - xn_hat[n, d])**2
  return lx
```

```python
def yn_hat(M_nk, w, y):
  import numpy as np
  N, K = M_nk.shape
  y_hat = np.dot(M_nk, w)
  return y_hat
```

```python
def ly(y_hat, y):
  import numpy as np
  log_loss_terms = -y * np.log(y_hat) - (1 - y) * np.log(1 - y_hat)
  ly = log_loss_terms.sum()
  return ly
```

```python
def L(params, africanamerican_data, caucasian_data, africanamerican_labels, caucasian_labels, K, A_z, A_x, A_y):
  import numpy as np
  africanamerican_samples, africanamerican_features = africanamerican_data.shape
  caucasian_samples, caucasian_features = caucasian_data.shape

  alpha_africanamerican = params[:africanamerican_features]
  alpha_caucasian = params[africanamerican_features : 2 * africanamerican_features]
  w = params[2 * africanamerican_features : (2 * africanamerican_features) + K]
  v = np.matrix(params[(2 * africanamerican_features) + K:]).reshape((K, africanamerican_features))

  dist_africanamerican = d(africanamerican_data, v, alpha_africanamerican)
  dist_caucasian = d(caucasian_data, v, alpha_caucasian)

  M_nk_africanamerican = M_nk(dist_africanamerican, K)
  M_nk_caucasian = M_nk(dist_caucasian, K)

  M_k_africanamerican = M_k(M_nk_africanamerican)
  M_k_caucasian = M_k(M_nk_caucasian)

  L_z = sum(abs(M_k_africanamerican[k] - M_k_caucasian[k]) for k in range(K))

  x_hat_africanamerican = xn_hat(africanamerican_data, M_nk_africanamerican, v)
  L_x_africanamerican = lx(africanamerican_data, M_nk_africanamerican, x_hat_africanamerican)
  x_hat_caucasian = xn_hat(caucasian_data, M_nk_caucasian, v)
  L_x_caucasian = lx(caucasian_data, M_nk_caucasian, x_hat_caucasian)
  L_x = L_x_africanamerican + L_x_caucasian

  y_hat_africanamerican = yn_hat(M_nk_africanamerican, w, africanamerican_labels)
  L_y_africanamerican = ly(y_hat_africanamerican, africanamerican_labels)
  y_hat_caucasian = yn_hat(M_nk_caucasian, w, caucasian_labels)
  L_y_caucasian = ly(y_hat_caucasian, caucasian_labels)
  L_y = L_y_africanamerican + L_y_caucasian

  loss = A_z * L_z + A_x * L_x + A_y * L_y

  return loss
```

```python
def cal_pred(params, D, K, africanamerican_dt, caucasian_dt, africanamerican_label, caucasian_label):

    alphas_africanamerican = params[:D]
    alphas_caucasian = params[D : 2 * D]
    weights = params[2 * D : (2 * D) + K]
    v_matrix = np.matrix(params[(2 * D) + K:]).reshape((K, D))

    dist_africanamerican = d(africanamerican_dt, v_matrix, alphas_africanamerican)
    dist_caucasian = d(caucasian_dt, v_matrix, alphas_caucasian)
    M_nk_africanamerican = M_nk(dist_africanamerican, K)
    M_nk_caucasian = M_nk(dist_caucasian, K)

    y_hat_africanamerican = yn_hat(M_nk_africanamerican, weights, africanamerican_label)
    likelihood_africanamerican = ly(y_hat_africanamerican, africanamerican_label)
    y_hat_caucasian = yn_hat(M_nk_caucasian, weights, caucasian_label)
    likelihood_caucasian = ly(y_hat_caucasian, caucasian_label)

    return y_hat_africanamerican, y_hat_caucasian
```

```python
# calculates the overall accuracy,
# accuracy for African-American group
# accuracy for Caucasian group
# and calibration of the model

def cal_calibr(y_pred_africanamerican, y_pred_caucasian, y_africanamerican_labels, y_caucasian_labels):

    converted_y_pred_africanamerican = [1 if pred >= 0.5 else 0 for pred in y_pred_africanamerican]
    converted_y_pred_caucasian = [1 if pred >= 0.5 else 0 for pred in y_pred_caucasian]

    accuracy_africanamerican = accuracy_score(y_africanamerican_labels, converted_y_pred_africanamerican)
    accuracy_caucasian = accuracy_score(y_caucasian_labels, converted_y_pred_caucasian)

    combined_labels = y_africanamerican_labels.append(y_caucasian_labels)
    combined_predictions = np.concatenate((converted_y_pred_africanamerican, converted_y_pred_caucasian), axis=0)
    overall_accuracy = accuracy_score(combined_labels, combined_predictions)
    calibration = abs(accuracy_africanamerican - accuracy_caucasian)

    table = [["Overall",overall_accuracy],["African-American",accuracy_africanamerican],
            ["Nonsensitive",accuracy_caucasian],
             ["Calibration",abs(accuracy_africanamerican - accuracy_caucasian)]]
    from tabulate import tabulate
    print(tabulate(table, headers = ["Race","Accuracy"],tablefmt = "simple"))
```

```python
def LFR(training_data, val_data, y_name, variable_name, K, A_z, A_x, A_y):
    africanamerican_training = training_data[training_data[variable_name] == 0]
    caucasian_training = training_data[training_data[variable_name] == 1]
    africanamerican_validation = val_data[val_data[variable_name] == 0]
    caucasian_validation = val_data[val_data[variable_name] == 1]

    for dataset in [africanamerican_training, africanamerican_validation, caucasian_training, caucasian_validation]:
        dataset.drop(columns=[variable_name], inplace=True)

    y_africanamerican_training = africanamerican_training.pop(y_name)
    y_africanamerican_validation = africanamerican_validation.pop(y_name)
    y_caucasian_training = caucasian_training.pop(y_name)
    y_caucasian_validation = caucasian_validation.pop(y_name)

    # Generating random alpha and w
    alpha_africanamerican = np.random.dirichlet(np.ones(africanamerican_training.shape[1]), size=1).flatten()
    alpha_caucasian = np.random.dirichlet(np.ones(caucasian_training.shape[1]), size=1).flatten()
    w = np.random.dirichlet(np.ones(K), size=1).flatten()
    v = np.random.random((K, africanamerican_training.shape[1]))
    initial_parameters = np.concatenate([alpha_africanamerican, alpha_caucasian, w, v.flatten()])

    bounds = [(0, 1)] * (africanamerican_training.shape[1] + caucasian_training.shape[1] + K) + [(None, None)] * (K * africanamerican_training.shape[1])
    optimal_params, min_L, _ = optim.fmin_l_bfgs_b(L, x0=initial_parameters, epsilon=1e-5,
                                                   args=(africanamerican_training, caucasian_training, y_africanamerican_training,
                                                         y_caucasian_training, K, A_z, A_x, A_y),
                                                   bounds=bounds, approx_grad=True,
                                                   maxfun=150000, maxiter=150000)

    y_hat_africanamerican_train, y_hat_caucasian_train = cal_pred(optimal_params, africanamerican_training.shape[1], K, africanamerican_training,
                                                                 caucasian_training, y_africanamerican_training, y_caucasian_training)
    print("Training accuracy:")
    cal_calibr(y_hat_africanamerican_train, y_hat_caucasian_train, y_africanamerican_training, y_caucasian_training)

    y_hat_africanamerican_val, y_hat_caucasian_val = cal_pred(optimal_params, africanamerican_validation.shape[1], K, africanamerican_validation,
                                                             caucasian_validation, y_africanamerican_validation, y_caucasian_validation)
    print("Validation accuracy:")
    cal_calibr(y_hat_africanamerican_val, y_hat_caucasian_val, y_africanamerican_validation, y_caucasian_validation)

    return optimal_params
```

```python
start = time.time()
para_test = LFR(data_train, data_val, 'two_year_recid', 'race', 10, 0.3, 0.3, 0.4)
end = time.time()
print( f"The training time: {end-start}")
```

```python
# Test the LFR model

import time

def prepare_data(dataset, race_id, outcome_col='two_year_recid'):
    filtered_data = dataset[dataset['race'] == race_id].drop(columns=['race'])
    outcome = filtered_data[outcome_col]
    return filtered_data.drop(columns=[outcome_col]), outcome

african_american_data, y_african_american = prepare_data(data_test, 0)
caucasian_data, y_caucasian = prepare_data(data_test, 1)

# Runtime of testing LFR model
start_time = time.time()
y_hat_african_american, y_hat_caucasian = cal_pred(para_test, african_american_data.shape[1], 10, african_american_data, caucasian_data, y_african_american, y_caucasian)
end_time = time.time()

print(f"Runtime of testing LFR model: {end_time - start_time}")

# Calibration
cal_calibr(y_hat_african_american, y_hat_caucasian, y_african_american, y_caucasian)
```
